```python
#Analisis de variables categoricas
import matplotlib.pyplot as plt
import seaborn as sns

def graficar_categoricas(df1):
    categoricas = df1.select_dtypes(include=['object', 'category'])
    n_cols = 2
    n_rows = (len(categoricas.columns) + 1) // n_cols
    
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(14, n_rows * 5))
    axes = axes.flatten()

    for ax, col in zip(axes, categoricas.columns):
        categoricas[col].value_counts().plot(kind='bar', ax=ax, color='skyblue')
        ax.set_title(f"Distribución de: {col}")

    plt.tight_layout()
    plt.show()

graficar_categoricas(df1)
```

```python
#Analisis de variables numericas
def graficar_numericas(df1):
    numericas = df1.select_dtypes(include=['int64', 'float64'])
    n_cols = 2
    n_rows = (len(numericas.columns) + 1) // n_cols

    fig, axes = plt.subplots(n_rows * 2, n_cols, figsize=(14, n_rows * 10))
    axes = axes.flatten()

    for idx, col in enumerate(numericas.columns):
        
        numericas[col].value_counts().plot(kind='bar', ax=axes[idx*2], color='skyblue')
        axes[idx*2].set_title(f"Distribución de: {col}")

        
        numericas[col].plot(kind='box', ax=axes[idx*2 + 1], color='lightgreen')
        axes[idx*2 + 1].set_title(f"Boxplot de: {col}")

    plt.tight_layout()
    plt.show()


graficar_numericas(df1)
```

```python
#Funcion para factorizar df2
def factorizar_categoricas(df2):
    
    df2_factorizado = df2.copy()
    
    for col in df2_factorizado.select_dtypes(include=['object', 'category']).columns:
        df2_factorizado[col], _ = pd.factorize(df2_factorizado[col])
    
    return df2_factorizado
```

```python
df2_factorizado = factorizar_categoricas(df2)
```

```python
def calcular_correlacion(df, target):
    
    if target not in df2_factorizado.columns:
        raise ValueError(f"La columna '{target}' no existe en el DataFrame.")
    
    numericas = df2_factorizado.select_dtypes(include=['int64', 'float64'])
    
    
    correlaciones = numericas.corr()[target].drop(target)  # Excluye la autocorrelación con la misma variable
    
   
    correlaciones = correlaciones.sort_values(ascending=False)
    
    return correlaciones
```

```python
correlaciones = calcular_correlacion(df2_factorizado, 'help_finc_leg')
print(correlaciones)
```
